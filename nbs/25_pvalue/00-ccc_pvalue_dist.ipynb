# Description

Tests whether the distribution of pvalues generated by CCC is the expected one (uniform).

# Modules loading

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from ccc.coef import ccc
```

# Generate random data

```python
data = np.random.rand(100, 1000)
```

```python
data.shape
```

# Run CCC

```python
res = ccc(data, n_jobs=20, pvalue_n_perms=1000, use_ari_numba=True)
```

# Plots

```python
plt.hist(res[1], bins=10, edgecolor="k")  # Adjust the number of bins as needed
plt.title("Distribution of Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
```

```python
sns.histplot(res[1], kde=True, color="blue")
plt.title("Distribution of Values")
plt.xlabel("Value")
plt.ylabel("Density")
```

# KS

```python
stats.ks_1samp(res[1], stats.uniform.cdf)
```
